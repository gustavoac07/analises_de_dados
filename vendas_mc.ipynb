```python
#Importando a biblioteca pandas

import pandas as pd
```

```python
# carregando os dados em data frame

df_vendas_mc = pd.read_csv('/content/vendas_linha_petshop_2020_2024.csv', encoding='latin1', sep=';')
```

```python
# Lendo data frame.

df_vendas_mc.head()
```

```python
#Consultando as informações do tipo dos dados

df_vendas_mc.info()
```

```python
# Quantas linhas e colunas existem no data frame

df_vendas_mc.shape
```

```python
#consultando o nome das colunas

df_vendas_mc.columns
```

```python
# consultando numero de dados nulos

df_vendas_mc.isnull().sum()
```

```python
# consultando se existe dados duplicados

df_vendas_mc.duplicated().sum()
```

```python
# excluindo os dados nulos e dando um novo nome para o data frame tratado

df_vendas_mc_tratado = df_vendas_mc.dropna()
```

```python
# consultando novamente se existe dados nulos

df_vendas_mc_tratado.isnull().sum()
```

```python
# lendo as 10 primeiras linhas do data frame

df_vendas_mc_tratado.head(10)
```

```python
# valores vazios na coluna 'quantidade' por 0
df_vendas_mc_tratado['quantidade'].fillna(0, inplace=True)
```

```python

#  linhas duplicadas
duplicados = df_vendas_mc_tratado[df_vendas_mc_tratado.duplicated()]

# mostrar linhas duplicadas
print(duplicados)
```

```python
# Convertendo a coluna 'quantidade' para numérico, forçando a conversão de erros para NaN
df_vendas_mc_tratado['quantidade'] = pd.to_numeric(df_vendas_mc_tratado['quantidade'], errors='coerce')

# consultando a conversão
print(df_vendas_mc_tratado['quantidade'].dtype)
```

```python
df_vendas_mc_tratado.head(10)
```

```python
# consultando se existe dados nulos após a normalização dos dados:

df_vendas_mc_tratado.isnull().sum()
```

```python
#novamente excluindo os dados nulos

df_vendas_mc_tratado.dropna(inplace=True)
```

```python
#consulta após a exclusão

df_vendas_mc_tratado.isnull().sum()
```

```python
# Tratamento dos dados consultando os valores outliers

# Calcular quartis e IQR
q1 = df_vendas_mc_tratado['quantidade'].quantile(0.25)
q3 = df_vendas_mc_tratado['quantidade'].quantile(0.75)
iqr = q3 - q1

#  limites para outliers
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

# Identificar outliers
outliers_iqr = df_vendas_mc_tratado[(df_vendas_mc_tratado['quantidade'] < limite_inferior) | (df_vendas_mc_tratado['quantidade'] > limite_superior)]

print("Outliers identificados com base no IQR:")
outliers_iqr.head(20)
```

```python
# Calculando média e desvio padrão
media = df_vendas_mc_tratado['quantidade'].mean()
desvio_padrao = df_vendas_mc_tratado['quantidade'].std()

# Definindo os limites para outliers
limite_inferior = media - 3 * desvio_padrao
limite_superior = media + 3 * desvio_padrao

# Identificando outliers
outliers_desvio = df_vendas_mc_tratado[(df_vendas_mc_tratado['quantidade'] < limite_inferior) | (df_vendas_mc_tratado['quantidade'] > limite_superior)]

print("Outliers identificados com base no desvio padrão:")
outliers_desvio.head(20)
```

```python
import matplotlib.pyplot as plt

# boxplot para consultar outliers
plt.boxplot(outliers_desvio['quantidade'].dropna())
plt.title('Boxplot de Quantidade')
plt.ylabel('Quantidade')
plt.show()
```

```python
# mantendo apenas valores positivos em 'quantidade'
df_positivos = df_vendas_mc_tratado[df_vendas_mc_tratado['quantidade'] >= 0]

# consultando o DataFrame
print(df_positivos)
```

```python
# DataFrame original para excluir valores negativos
df_vendas_mc_final = df_vendas_mc_tratado[df_vendas_mc_tratado['quantidade'] >= 0]

# Consultando o DataFrame
print(df_vendas_mc_final)
```

```python
# Consultando a tabela final após as normalização

df_vendas_mc_final.head(10)
```

```python
# excluindo o 'R$' e substituir vírgulas por pontos
df_vendas_mc_final['valor_total_bruto'] = df_vendas_mc_final['valor_total_bruto'].str.replace('R\$', '', regex=True)
df_vendas_mc_final['valor_total_bruto'] = df_vendas_mc_final['valor_total_bruto'].str.replace(',', '.')

# Convertendo para numérico
df_vendas_mc_final['valor_total_bruto'] = pd.to_numeric(df_vendas_mc_final['valor_total_bruto'], errors='coerce')
```

```python
# excluindo o 'R$' e substituir vírgulas por pontos
df_vendas_mc_final['valor_comissao'] = df_vendas_mc_final['valor_comissao'].str.replace('R\$', '', regex=True)
df_vendas_mc_final['valor_comissao'] = df_vendas_mc_final['valor_comissao'].str.replace(',', '.')

# Convertendo para numérico
df_vendas_mc_final['valor_comissao'] = pd.to_numeric(df_vendas_mc_final['valor_comissao'], errors='coerce')
```

```python
# excluindo o 'R$' e substituir vírgulas por pontos
df_vendas_mc_final['lucro_liquido'] = df_vendas_mc_final['lucro_liquido'].str.replace('R\$', '', regex=True)
df_vendas_mc_final['lucro_liquido'] = df_vendas_mc_final['lucro_liquido'].str.replace(',', '.')

# Convertendo para numérico
df_vendas_mc_final['lucro_liquido'] = pd.to_numeric(df_vendas_mc_final['lucro_liquido'], errors='coerce')
```

```python
# convertendo texto para DATA

df_vendas_mc_final['data'] = pd.to_datetime(df_vendas_mc_final['data'], errors='coerce')
```

```python
# excluindo o 'R$' e substituir vírgulas por pontos
df_vendas_mc_final['valor'] = df_vendas_mc_final['valor'].str.replace('R\$', '', regex=True)
df_vendas_mc_final['valor'] = df_vendas_mc_final['valor'].str.replace(',', '.')

# Convertendo para numérico
df_vendas_mc_final['valor'] = pd.to_numeric(df_vendas_mc_final['valor'], errors='coerce')

df_vendas_mc_final.head(10)

```
